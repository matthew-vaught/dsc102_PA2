```python
import json

def check_match_and_sorted(stu_dict,gt_dict):
    # check if all keys match
    bool1 = sorted(list(stu_dict.keys())) == sorted(list(gt_dict.keys()))
    # check if all values match
    bool2 = sorted(list(stu_dict.values())) == sorted(list(gt_dict.values()))
    # check if values are sorted
    bool3 = True
    maxsofar=1e9
    for k in stu_dict:
        if stu_dict[k] > maxsofar:
            bool3 = False
            break
        maxsofar = stu_dict[k]
    return bool1 and bool2 and bool3


def calculate_points():
    global_timeout = 1800
    tolerance = 0.01

    with open("ground_truth_results_PA1.json",'r') as json_file: 
        gt_solutions = json.load(json_file)
        
    with open("my_results_PA1.json",'r') as json_file: 
        student_answer = json.load(json_file)
        
    total_pts = 0
    for k in gt_solutions:
        if k == 'q1':
            curkey_flag=1
            for nestedk in gt_solutions[k]:
                if abs(gt_solutions[k][nestedk] - student_answer[k][nestedk]) > (gt_solutions[k][nestedk] * tolerance):
                    curkey_flag = 0
                    print(f"q1 doesnt match")
                    break              
            if curkey_flag: total_pts += 6
        elif k == 'q2':
            curkey_flag=1
            for nestedk in gt_solutions[k]:
                if abs(gt_solutions[k][nestedk] - student_answer[k][nestedk]) > (gt_solutions[k][nestedk]* tolerance):
                    curkey_flag = 0
                    print(f"q2 doesnt match")
                    break            
            if curkey_flag: total_pts += 6       
        elif k == 'q3':
            if gt_solutions[k] == student_answer[k]: 
                total_pts += 14
            else:
                print(f"q3 doesnt match")
        elif k == 'q4':
            subpoints=0
            for nestedk in ["mean", "std", "median", "min", "max"]:
                if nestedk=="median":
                    subpoints += 2.4
                    continue
                if abs(gt_solutions[k][nestedk] - student_answer[k][nestedk]) <= (gt_solutions[k][nestedk] * tolerance):
                    subpoints += 2.4
                else:
                    print(f"q4 doesnt match")
            total_pts += subpoints         
        elif k == 'q5':  
            if check_match_and_sorted(student_answer[k],gt_solutions[k]): 
                total_pts += 14
            else:
                print(f"q5 doesnt match")
        elif k == 'q6':
            if gt_solutions[k] == student_answer[k]: 
                total_pts += 14
            else:
                print(f"q6 doesnt match")
        elif k == 'q7':
            if gt_solutions[k] == student_answer[k]: 
                total_pts += 14
            else:
                print(f"q7 doesnt match")
            
    print(f"POINTS = {total_pts}")
    print(f"RUNTIME = {student_answer['runtime']}")
    
    return total_pts

calculate_points()
```
